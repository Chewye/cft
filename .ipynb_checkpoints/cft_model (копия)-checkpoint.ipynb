Оставим значимые признаки, которые все являются категориальными и закодируем их ванхотингом

```python
products_prep = products[['level_1', 'level_2', 'level_3', 'level_4', 'segment_id', 'is_own_trademark']]
```

```python
products_prep.shape
```

```python
x = products_prep.to_numpy()
x = x.reshape((-1, 6))
```

```python
x
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.decomposition import PCA


n_components = 100
one_hot = OneHotEncoder()
pca = PCA(n_components=n_components)
decod_prod = one_hot.fit_transform(x)
```

```python
decod_prod.toarray()
```

```python
prod_array = pca.fit_transform(decod_prod.toarray())
```

```python
prod_array = pd.DataFrame(prod_array)
```

```python
prod_array['product_id'] = products['product_id']
prod_array.head()
```

```python
train_purch
```

```python
result = np.zeros([client.shape[0], n_components])

for cli in tqdm(range(client.shape[0])):    
    for prod in train_purch[train_purch['client_id'] == client['client_id'].iloc[cli]]['product_id']:
        #prod_id = train_purch[train_purch['client_id'] == client['client_id'].iloc[cli]]['product_id']
        arr = prod_array[prod_array['product_id'] == prod].to_numpy()[:, :n_components].astype(float)
        result[cli] += arr[0]       
        
```

```python
result
```

```python
result.shape
```

```python
res_train_df = pd.DataFrame(result)
```

```python
res_train_df.to_csv('res_train_df.csv')
```
